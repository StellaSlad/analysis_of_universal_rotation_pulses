# Import Required Libraries

````python
# Import Required Libraries
import os
import numpy as np
import matplotlib.pyplot as plt
import lib.calculations as c
import lib.plot_functions as pf
````

# Set Parameters

````python
# Pulse directory
DIRECTORY = 'pulses/'
# Timestep - temporal resolution of the pulse shape
TIMESTEP = 2.0e-6
# Ideal rotation axis
IDEAL_AXIS = np.array([1, 0, 0]) 
# Ideal rotation angle
ANGLE = 90 
# Frequency offset range in Hz
OFFSRANGE = 10000 
# Number of frequency offsets
N_OFFSETS = 101 
# Propagator for 90x pulse
U_DESIRED = np.array([[np.sqrt(2)/2, np.sqrt(2)*1j/2], [-np.sqrt(2)/2, np.sqrt(2)*1j/2]])  
# Offsets array
OFFSETS = np.linspace(-OFFSRANGE/2, OFFSRANGE/2, N_OFFSETS)
````

 ## How to Use `main.py`

To execute the entire analysis, you can simply run the `main.py` script. This script will ensure the necessary directories exist, change the working directory, get the pulse filenames, and perform all the calculations and plotting.

To execute the main function, run:

```python\n,

if __name__ == "__main__":
    main()
```

The `main` function calls all the necessary functions in the correct order. If you change the order, please make sure that the plotting functions are called, after all the calculations have been performed.

The following demo demonstrates the individual functions, while analysing a BURBOP-90x pulse.

# Calculate Effective Propagator

````python
# Pulse name
file_name = "pulse1400"

# Change to the pulses directory
os.chdir(DIRECTORY)

# Calculate Effective Propagator
c.calculate_effective_propagator(file_name, OFFSRANGE, N_OFFSETS)
````

# Calculate and Plot Rotation Axis

````python
# Calculate Rotation Axis
Lxx, Lyy, Lzz, rx, ry, rz = c.calculate_rotation_axis(file_name, OFFSETS, N_OFFSETS)

# Plot Rotation Axis
pf.plot_rotation_axis(file_name, OFFSETS, rx, ry, rz, Lxx, Lyy, Lzz)
````

# Calculate Final States and Quality Factor of the pulse


````python
# Calculate Final States
c.calculate_final_states(file_name, N_OFFSETS)

# Calculate Quality Factor
c.calculate_quality_factor(file_name, N_OFFSETS, U_DESIRED)
````

````python
# Plot Transfer Efficiency
pf.plot_transfer_efficiency(file_name, OFFSETS)
# Placeholder for figures
plt.show()
````
